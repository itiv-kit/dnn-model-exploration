```python
# scale generation (-1.5 to +1.5)

from codecs import decode
import struct
import ctypes

def int_to_bytes(n, length):
	return decode('%%0%dx' % (length << 1) % n, 'hex') [-length:]
def bin_to_float(b):
	bf = int_to_bytes(int(b, 2), 4)
	return struct.unpack('>f', bf)[0]

def float_to_bfloat16(value):
	value_fp32_bin = bin(ctypes.c_uint32.from_buffer(ctypes.c_float(value)).value)
	print(value_fp32_bin)
	value_fp32_bin = value_fp32_bin[:-16] + '0000' + '0000' + '0000' + '0000'  
	print(value_fp32_bin[:-16])

	return bin_to_float(value_fp32_bin)
    
def scale_gen(bits):
	""" calculates the corresponding scale for the width of -1.5 to +1.5 values and respecting the bitwidth"""
	print('bits:' , bits)
	scale = 1.5 / pow(2,bits-1)
	print('real scale: ', scale )
	scale_bfloat16 = float_to_bfloat16(scale)

	print('real inv_scale: ', 1/scale )
	inv_scale_bfloat16 = float_to_bfloat16(1/scale)

	print('bfloat16 scale', scale_bfloat16)
	print('bfloat16 inv_scale', inv_scale_bfloat16)
	print('\n\n')
	return [scale_bfloat16, inv_scale_bfloat16]
 
for i in range(2,17):
	scale_gen(i)

```
